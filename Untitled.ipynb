```python
import numpy as np
import pandas as pd
import csv
import os
```

```python
B11 = pd.read_csv('b11-f12-uvis-overlap_star.csv')
#print(B9['RA'])
```

```python
data = np.zeros(len(B11['RA'])+1)
for i in range(len(data)):
    data[i] = 56511
print(data)
print(len(data))
print(len(B11['RA']))
```

```python
with open('b11-f12-uvis-overlap_star.csv','r') as csvFile:  #此处的csv是源表，即想要写入的表
    rows = csv.reader(csvFile)
    #print(rows)
    with open('b11-f12-uvis-overlap_star_mjd.csv','w') as f: #这里的csv则是最后输出得到的新表
        writer = csv.writer(f)
        i = 0
        for row in rows:
            row.append(data[i])
            #print(i)
            i = i + 1
            writer.writerow(row)
```

```python
inputfile_csv_1="b9-f05-uvis-overlap_star_mjd.csv"
inputfile_csv_2="b11-f12-uvis-overlap_star_mjd.csv"
outputfile="B9-B11-uvis-overlap_star.csv"
csv_1=pd.read_csv(inputfile_csv_1)
csv_2=pd.read_csv(inputfile_csv_2)
out_csv=pd.concat([csv_1,csv_2],axis=0)
out_csv.to_csv(outputfile,index=False)
```
